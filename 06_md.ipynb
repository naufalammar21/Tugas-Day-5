Loop with Range

Buatlah script untuk mencetak dengan rentang angka tertentu.
Contoh

Input pertama: 4
Input kedua: 8
Output di browser: 4, 5, 6, 7, 8


```python
angka1 = int(input('input angka pertama : '))
angka2 = int(input('input angka kedua : '))

for i in range (angka1,angka2):
  print(i)
```
